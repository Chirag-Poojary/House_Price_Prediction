<h2><b>Importing Libraries</b></h2>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<h2><b>Load Dataset</b></h2>

```python
data = pd.read_csv('../backend/datasets/boston_dataset.csv')
df = pd.DataFrame(data)
```

<h2><b>Understanding Columns</b></h2>

```python
plt.figure(figsize=(10,8))
sns.boxplot(df)
plt.title('Boxplot of data')
plt.show()
```

```python
#Using Log Transformation for columns CRIM & TAX
df['CRIM'] = np.log1p(df['CRIM'])
df['TAX'] = np.log1p(df['TAX'])
df['MEDV'] = df[df['MEDV'] < 50]['MEDV']
```
