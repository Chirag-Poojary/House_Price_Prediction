```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../backend/datasets/boston_cleaned.csv')
```

```python
plt.figure(figsize=(8,6))
sns.kdeplot(df.MEDV)
sns.histplot(df['MEDV'], bins=30, kde=True, stat="density")
plt.xlim(0,60)
plt.title('Output Distribution')
plt.show()
# Target is Right Skewed
```

```python
sns.pairplot(df)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x='RM', y='MEDV', data=df)
plt.title('RM vs MEDV')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x='CHAS', y='MEDV', data=df)
plt.title('CHAS vs MEDV')
plt.show()
```

```python
price_bins = [0, 15, 25, 35, 60]
price_labels = ['Low', 'Medium', 'High', 'Very High']
df['price_group'] = pd.cut(df['MEDV'], bins=price_bins, labels=price_labels)
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='price_group', y='RM', data=df)
plt.title("Boxplot of RM across Price Buckets")
plt.xlabel("Price Group")
plt.ylabel("Average Number of Rooms (RM)")
plt.show()
```

```python
plt.figure(figsize=(8,6))
for group in df['price_group'].unique():
    sns.kdeplot(df[df['price_group'] == group]['RM'], label=group, linewidth=2)
plt.title("KDE of RM by Price Group")
plt.xlabel("Average Number of Rooms (RM)")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='price_group',y='LSTAT',data=df)
plt.title("Boxplot of LSTAT across Price Buckets")
plt.xlabel("Price Group")
plt.ylabel("% lower status of the population")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='price_group', y='LSTAT', data=df, showfliers=False)
sns.stripplot(x='price_group', y='LSTAT', data=df, color='black', alpha=0.4,
 jitter=True)
plt.title("Distribution of LSTAT across Price Buckets")
plt.xlabel("Price Group")
plt.ylabel("% Lower Status of the Population")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='price_group',y='PTRATIO',data=df)
plt.title("Boxplot of PTRATIO across Price Buckets")
plt.xlabel("Price Group")
plt.ylabel("Pupil-Teacher Ratio")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='price_group',y='NOX',data=df)
plt.title("Boxplot of NOX across Price Buckets")
plt.xlabel("Price Group")
plt.ylabel("Nitric oxide concentration")
plt.show()
```

```python
plt.figure(figsize=(8,6))
for group in df['price_group'].unique():
    sns.kdeplot(df[df['price_group'] == group]['NOX'], label=group, linewidth=2)
plt.title("KDE of NOX by Price Group")
plt.xlabel("Nitric oxide concentration")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='price_group',y='DIS',data=df)
plt.title("Boxplot of DIS across Price Buckets")
plt.xlabel("Price Group")
plt.ylabel("Distance to employment centers")
plt.show()
```
